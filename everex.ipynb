```python
import os
import struct
import Image
import scipy
import scipy.misc
import scipy.cluster
from IPython.display import Image as IpyImage

%timeit

NUM_CLUSTERS = 12
file_name = "temp"
print 'reading image'
im = Image.open("everex.png")
# im = im.resize((300, 300))      optional, to reduce time
ar = scipy.misc.fromimage(im)
shape = ar.shape 

print "this is shape", shape
ar = ar.reshape(scipy.product(shape[:1]), shape[1])

print 'finding clusters'
codes, dist = scipy.cluster.vq.kmeans(ar, NUM_CLUSTERS)
print 'cluster centres:\n', codes

vecs, dist = scipy.cluster.vq.vq(ar, codes)         # assign codes
counts, bins = scipy.histogram(vecs, len(codes))    # count occurrences

print "all bins:", bins
for each_color in codes:
    print ''.join(chr(c) for c in each_color).encode('hex')

index_max = scipy.argmax(counts)                    # find most frequent
peak = codes[index_max]
print "peak:", peak
colour = ''.join(chr(c) for c in peak).encode('hex')
print 'most frequent is %s (#%s)' % (peak, colour)
c = ar.copy()
for i, code in enumerate(codes):
    c[scipy.r_[scipy.where(vecs==i)],:] = code
scipy.misc.imsave("clusters_image/clusters_"+file_name+".png", c.reshape(*shape))
print 'saved clustered image'
print "This is how it will get printed"
IpyImage(filename="clusters_image/clusters_"+file_name+".png")
```
